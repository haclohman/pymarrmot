```python
# required imports
import os
import pandas as pd
import holoviews as hv
from bokeh.plotting import show

# directories
input_temp_dir = "C:\\Data\\nwm30_forcing\\temperature"
output_temp_dir = "C:\\Users\\ssheeder\\Repos\\pymarrmot\\forcing\\pymarrmot\\temp"
combined_forcing_dir = "C:\\Users\\ssheeder\\Repos\\pymarrmot\\forcing\\pymarrmot\\combined_forcing"
usgs_dir = "C:\\Users\\ssheeder\\Repos\\pymarrmot\\forcing\\pymarrmot\\usgs"

# test parameters
usgs_name = "usgsbasin-02111180"
start_date = "2014-11-01"
end_date = "2014-11-30"
```

```python
# Katie identified a problem with November 2014 data in the combined pymarrmot forcing file
# checking the temperature data for that month, prior to the merge with other variables
processed_temp = os.path.join(output_temp_dir, usgs_name + ".parquet")
df_processed = pd.read_parquet(processed_temp)
df_processed = df_processed[(df_processed["value_time"] >= start_date) & (df_processed["value_time"] <= end_date)]
hv.extension('bokeh')
hv_curve = hv.Curve(df_processed, 'value_time', 'temp_c')
show(hv.render(hv_curve))
```

```python
#now looking at the temperature data in the combined forcing file
#no data after Nov 18 - so something wrong with merge statement in combine_forcing.py

test2_processed = os.path.join(combined_forcing_dir, usgs_name + "_combined.parquet")
df2_processed = pd.read_parquet(test2_processed)
df2_processed = df2_processed[(df2_processed["value_time"] >= start_date) & (df2_processed["value_time"] <= end_date)]
hv.extension('bokeh')
hv_curve = hv.Curve(df2_processed, 'value_time', 'temp_c')
show(hv.render(hv_curve))
```

```python
#now looking at the usgs data in the combined forcing file
#no data after Nov 18 - so something wrong with merge statement in combine_forcing.py

test3_processed = os.path.join(combined_forcing_dir, usgs_name + "_combined.parquet")
df3_processed = pd.read_parquet(test3_processed)
df3_processed = df3_processed[(df3_processed["value_time"] >= start_date) & (df2_processed["value_time"] <= end_date)]
hv.extension('bokeh')
hv_curve = hv.Curve(df3_processed, 'value_time', 'discharge_mm')
show(hv.render(hv_curve))
```

```python
# Now looking at the USGS file prior to merge

processed_usgs = os.path.join(usgs_dir, usgs_name + ".parquet")
df_usgs = pd.read_parquet(processed_usgs)
df_usgs = df_usgs[(df_usgs["value_time"] >= start_date) & (df_usgs["value_time"] <= end_date)]
df_usgs.head()

hv.extension('bokeh')
hv_curve = hv.Curve(df_usgs, 'value_time', 'discharge_mm')
show(hv.render(hv_curve))
```

```python
#trying to merge usgs with the temperature information
df_combined = df_processed.merge(df_usgs, on=['value_time', 'value_time'], how='left')

hv.extension('bokeh')
hv_curve = hv.Curve(df_combined, 'value_time', 'temp_c') #temp_c or discharge_mm
show(hv.render(hv_curve))
```
