```python
import os
import pandas as pd
```

```python
combined_forcing_dir = "C:\\Users\\ssheeder\\Repos\\pymarrmot\\forcing\\pymarrmot\\combined_forcing"
output_dir = "C:\\Users\\ssheeder\\Repos\\pymarrmot\\forcing\\pymarrmot\\combined_forcing\\12_year"

target_date = pd.to_datetime('2010-01-01')
```

```python
#for each parquet file in the pymarrmot/combined_forcing directory, open file and remove rows prior to 2010-01-01
forcing_files = os.listdir(combined_forcing_dir)
for file in forcing_files:
    if file.endswith('.parquet'):
        df = pd.read_parquet(os.path.join(combined_forcing_dir, file))
        df_trimmed = df[df['value_time'] >= target_date]
        df_trimmed.to_parquet(os.path.join(output_dir, file))
```
