```python
# required imports
import os
import pandas as pd

# file to process USGS observation data for use in PyMARRMoT
usgs_data = "C:\\Data\\usgs_obs.parquet"
usgs_area = "C:\\Data\\usgs_basin_attr_drainage_area.all.parquet"
```

```python
# ----------------------------USGS Data Processing----------------------------
df_basin_area = pd.read_parquet(usgs_area)
df_basin_area.head()
```

```python
df_usgs_data = pd.read_parquet(usgs_data)
df_usgs_data.head()

```

```python
unique_locations = df_usgs_data['location_id'].unique()
unique_locations    
```

```python
from dataretrieval import nwis as nwis
results = pd.DataFrame()
for id in unique_locations:
    station_id = id[5:]
    site_info = nwis.get_record(sites=station_id, service='site')
    results = pd.concat([site_info, results], ignore_index=True)

results.head()
```

```python
results.to_csv('usgs_station_summary')
```
